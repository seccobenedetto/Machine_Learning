## Image segmentation with k-means

In this laboratory we will use the k-means algorithm to cluster a dataset of 3D points. We will apply **K-means** to the problem of image compression and image segmentation. The main idea is to apply k-means to the colors of the pixels of an image to select the k most representative colors. Then, we will replace each pixel color with the closest representative color. This will allow us to reduce the number of colors in the image and compress it. A color is a vector of 3 values (R,G,B) that represent the amount of red, green and blue in the color; this implies that each pixel is a point in a 3D space.

```python
def img_plot(img: np.ndarray, title: str = None) -> None:
    """
    Plot an image
    :param img: image to plot
    :param title: title of the plot
    """
    
    plt.figure()
    plt.imshow(img)
    plt.axis('off')
    if title is not None:
        plt.title(f'{title}: {img.shape}')
    plt.tight_layout()
    plt.show()
    plt.close()
```

```python
def scatter_plot(data: np.ndarray, clusters: np.ndarray = None, centers: np.ndarray = None, title: str = None) -> None:
    """
    Plot a scatter plot of the data
    :param data: data to plot
    :param clusters: cluster labels
    :param centers: cluster centers
    :param title: title of the plot
    """
    
    fig = plt.figure()
    axis = fig.add_subplot(1, 1, 1, projection="3d")
    axis.set_xlabel("Red")
    axis.set_ylabel("Green")
    axis.set_zlabel("Blue", rotation=90, labelpad=-1)
    if title is not None:
        plt.title(title)
    if clusters is None:
        axis.scatter(data[:,0], data[:,1], data[:,2], marker="o", c=data, s=5)
    else:
        axis.scatter(data[:,0], data[:,1], data[:,2], marker="o", c=clusters, s=1, cmap='viridis', zorder=0, alpha=0.5 )
    if centers is not None:
        axis.scatter(centers[:,0], centers[:,1], centers[:,2], c='red', s=400, zorder=10)
    plt.tight_layout()
    plt.show()
    plt.close()
```

```python
def scatter_plot_2d(x: np.ndarray, y: np.ndarray = None, centers: np.ndarray = None, title: str = None) -> None:
    """
    Plot a scatter plot of the data
    :param x: data to plot
    :param y: cluster labels
    :param centers: cluster centers
    :param title: title of the plot
    """

    fig = plt.figure()
    plt.scatter(x[:,0], x[:,1], c=y, marker="o", s=10, cmap='viridis')
    plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5)
    if title is not None:
        plt.title(title)
    plt.tight_layout()
    plt.show()
    plt.close()
```

```python
def error_plot(errors: np.ndarray, labels: np.ndarray = None) -> None:
    """
    Plot the errors over the iterations
    :param errors: errors to plot
    """
    
    if labels is None:
        plt.plot(errors[1:-1])
        plt.plot(errors[1:-1], 'ro')
    else:
        plt.plot(labels, errors)
        plt.plot(labels, errors, 'ro')
    plt.title('Error over iterations')
    plt.ylabel('Error')
    plt.xlabel('Iteration #')
    plt.grid()
    plt.tight_layout()
    plt.show()
    plt.close()
```

```python
def cluster_plot(labels: np.ndarray, x: np.ndarray, title: str = None) -> None:
    """
    Plot the clusters
    :param labels: cluster labels
    :param x: data
    :param title: title of the plot
    """
    
    # Black removed and is used for noise instead.
    unique_labels = set(labels)
    colors = [plt.cm.Spectral(each)
            for each in np.linspace(0, 1, len(unique_labels))]
    for k, col in zip(unique_labels, colors):
        if k == -1:
            # Black used for noise.
            col = [0, 0, 0, 1]

        class_member_mask = (labels == k)

        xy = x[class_member_mask ]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
                markeredgecolor='k', markersize=14)

        xy = x[class_member_mask ]
        plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
                markeredgecolor='k', markersize=6)

    if title is not None:
        plt.title(title)
    plt.tight_layout()
    plt.show()
```

Implement the k-means algorithm manually (**do not use the kmeans function of sklearn and do not download implementations from other web sources**). The inputs to the function is the set of vectors to be clustered and the number of clusters. The output must contain the clusters barycenters, a vector associating each data point to the corresponding cluster and the error (value of the cost function) at each iteration.
Additionally, fix a maximum number of iterations of the k-means algorithm (e.g., 50).

Be careful about the initalization, you can use some random points from the training set, or get random values but ensure they are in the proper range.
```python
import numpy.random as npr
import scipy.linalg as la

def kmeans(points: np.ndarray, k: int, max_iters: int = 50) -> tp.Tuple[np.ndarray, np.ndarray, list]:
    """
    Perform K-means clustering
    :param points: data points
    :param k: number of clusters
    :param max_iters: maximum number of iterations
    """

    # TODO: Generate random centroids
    # Hint: Use the correct means and standard deviation to ensure they represent the whole input space
    point_count = points.shape[0]
    coord_count = points.shape[1]
    ##################################################################################_my code from here
    # INITIALIZING CENTROIDS:
    means = np.mean(points, axis=0)
    devs =   np.std(points, axis=0)
    centroids = np.array( [  [ npr.normal(means[i], devs[i]) for i in range(coord_count) ] for j in range(k)  ]  ) # k centroids of 3 coordinates each
    ##################################################################################_my code til here
    
    clusters = np.zeros((point_count,))   # array con elenco dei cluster a cui apparterr ogni punto

    prev_error = 1e9+1
    error = [1e9]
    
    # Run the K-means algorithm: until the estimate of centroid is stable (or the maximum number of iterations is reached)
    iters = 0
    while (error[iters] != prev_error) and iters < max_iters:    # ciclo continua fino a quando l'errore non cambia tra due iterazioni consecutive 
        ##################################################################################_my code from here
        # Measure the distance to every centroid
        distance = np.zeros((point_count, k))     # each i-th row has the distance of i-th point with each of the k centroids
        for i_data in range(point_count):
            for i_centr in range(k):
                distance[i_data, i_centr] = la.norm( points[i_data,:] - centroids[i_centr, :] )
        # TODO Step 1: assign training data to closest centroid
        for i_data in range(point_count):
            clusters[i_data] = np.argmin(distance[i_data,:])
        # TODO Calculate the new centroids for every cluster:
        for i_clust in range(k):
            data_clust = points[ clusters==i_clust ]    # datas of the same cluster i_clust
            sum = np.sum( data_clust, axis=0 )          # sum over columns (=over same coordinates)
            avg = sum / len( data_clust )               # average of each coordinate
            centroids[i_clust,:] = avg                  # avg is an array of three elements
        ##################################################################################_my code til here            
        prev_error = error[iters]
        current_error = 0 
        ##################################################################################_my code from here
        # TODO Update the total cost estimate
        for i_data in range(point_count):
            for i_coo in range(coord_count):
                i_k = int( clusters[i_data] )    # cluster di appartenenza
                dist = points[i_data, i_coo] - centroids[i_k, i_coo]
                current_error += dist**2
        current_error /= point_count
        ##################################################################################_my code til here
        error.append(current_error)
        
        # Update the iteration counter
        iters += 1

    return centroids, clusters, error
```

## B) DBSCAN clustering

Now, let's try DBSCAN: implement the algorithm with $\varepsilon=0.1$ and $M=10$ (i.e., the minimum number of neighbors). The output should be a list of indices for all points, indicating which cluster each point belongs to. Outliers are in cluster 0, all other points have a positive index. For efficiency, we should compute a distance matrix first. Additionally, since DBSCAN is much more complex, we will try it on a smaller dataset

```python
# Load sample data
data = np.load("data/moon_data.npz")

# Extract data
x = data['X']
print(x)
labels_true = data['labels_true']
```

Nota: Se la distanza di un punto da s stesso fosse vista come zero (come dovrebbe essere), il punto verrebbe sempre contato come un vicino di s stesso (dando problemi al DBScan), quindi si pone la distanza da se stessi come un numero enorme.

```python
import numpy.random as npr

def distance_matrix(points):                                     # 'points'  una matrice Nx3
    point_count = points.shape[0]
    distances = np.zeros((point_count, point_count))             # 'distances'  una matrice NxN
    for i in range(point_count):
        array_points = np.asarray(points - points[i, :])         # Sottrae il punto i-esimo da tutti gli altri punti
        distances[i, :] = np.linalg.norm(array_points,axis=1)    # Calcola la norma L2 (distanza euclidea) per ciascuna riga
        distances[i,i] = 1e9                              # Imposta una distanza molto grande per evitare che un punto sia vicino a s stesso
    return distances    # returns a NxN matrix with each distance


def find_neighbors(distances, epsilon):                          # funzione che scandaglia tutti i punti cercando i vicini
    return np.where(distances < epsilon)[0]                      # avendo messo [0], la devo usare su array 1D al massimo


def dbscan(points, epsilon, M):
    point_count = points.shape[0]
    coord_count = points.shape[1]
    distances = distance_matrix(points)    # NxN matrix
    print(point_count)
    C = 0            # contatore dei cluster
    done = False     # flag che controlla se tutti i punti sono stati assegnati
    # The clusters vector contains the assigned cluster values: -1 means unassigned, 0 means outlier, clusters have >=1 values
    clusters = -np.ones(point_count)
    while (not done):
        ##################################################################################_my code from here
        # TODO While there are still unassigned points, iterate over them
        # For each unassigned point, check if it is an outlier, expand over its neighbors if it is not 
        if ( len( clusters[ clusters==-1 ] ) )  > 0:   # if there are yet unassigned points    
            idx_u = np.where( clusters == -1 )[0]      # indexes of unassigned points
            idx_x = npr.choice( idx_u )                # choose one random index of a point x
            idx_neigh = find_neighbors(distances[idx_x,:], epsilon)   # indici dei vicini ad x
            if len(idx_neigh) < M: clusters[idx_x] = 0                # assign to outliers if there are few neighbors
            else:
                C +=1
                clusters[idx_x] = C
                while ( len(idx_neigh)>0 ):
                    n = npr.choice(idx_neigh)    # choose random neighbor index
                    if (clusters[n] == -1 or clusters[n] == 0):   
                        clusters[n] = C       # if the neighbor was outlier or unassigned, it's now in cluster C
                        idx_neigh_n = find_neighbors(distances[n,:], epsilon)
                        idx_neigh = np.concatenate( [idx_neigh, idx_neigh_n ])     # extend list of neighbors to check
                    index_of_n_in_neigh_list = np.where(idx_neigh==n)[0]
                    idx_neigh = np.delete( idx_neigh, index_of_n_in_neigh_list )   # I remove n from the neighbors
        else: done = True   # in this way it exit the cycle
        ##################################################################################_my code til here
    return clusters
```
