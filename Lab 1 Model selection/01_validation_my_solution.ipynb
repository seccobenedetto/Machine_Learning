## Polynomial Classification on Signal to Noise Ratios

In this notebook we are going to explore the use of polynomial classification with polynomial regression. We are going to use the Numpy **polyfit** function, which performs polynomial regression.

Our use case is a communication problem: we have a set of measurements of the Signal to Noise Ratio (SNR), i.e., the quality of the communication link, in various positions. The SNR depends on two components: firstly, the noise level (which is a random variable that does not depend on position) and the signal attenuation (usually modeled as a polynomial function of the distance).

Our transmitter is in (0,0), and coordinates are in meters. In urban scenarios, the attenuation usually follows a third-degree polynomial, but it might be a fourth- or fifth-degree polynomial in more complex cases. How do we choose between different degrees? We will try with a maximum of **6**

(solution: original data were taken from 3rd grade polynomium)

## Helper functions

These functions will help us evaluate the results

```python
def compute_distance(x, y):
    # Compute the Euclidean distance from the origin
    distance = np.sqrt( x**2 + y**2 )        
    return distance

def fit(distance, SNR, degree):
    # memo: polyfit restituisce un array  con i coeff del polinomio, ordinati da quello di grado piÃ¹ alto a quello di grado 0. 
    return np.polyfit(distance, SNR, deg=degree)   

def predict(distance, poly_coeffs):            # Predict the SNR from a given model
    deg = len(poly_coeffs) - 1    # degree of polynom
    SNR = 0                     # prediction value to calculate
    for i in range(deg+1): SNR += poly_coeffs[i]*distance**(deg - i)
    return SNR

def evaluate(distance, SNR, poly_coeffs):
    # Compute the error of the polynomial fit on the chosen data
    N = len(distance)
    SNR_predict = predict(distance, poly_coeffs)
    mse = np.sum( (SNR_predict - SNR)**2 )/N
    return mse

def separate_test(distance, SNR, test_points):
    # Return a training set and a test set (the test_points parameter controls the number of test points).
    # The points are selected randomly manipulating indexes
    tot_idx = range(len(distance))
    idx_mix = npr.permutation(tot_idx)
    x_test = distance[ idx_mix[0:test_points] ]
    y_test = SNR[ idx_mix[0:test_points] ]
    x_train = distance[ idx_mix[test_points:] ]
    y_train = SNR[ idx_mix[test_points:] ]    
    return x_train, y_train, x_test, y_test
```

## A. K-fold cross-validation

In this case, x and y are the two coordinates, and the SNR is the thing we are trying to predict

Flusso della validazione:
* Dividi i dati in k fold.
* Per ogni grado di polinomio (da 0 a max_degree), e per ogni fold (da 0 a k):
    * Allena il modello (=fai fit) con i dati dei k-1 fold.
    * Valida il modello con il fold rimanente calcolando l'errore (MSE) tra le previsioni e i dati reali del fold di validazione.
* Per ogni grado di polinomio calcola la media degli errori (mean_mse) su tutti i fold.
* Alla fine, il grado che ha il miglior errore medio (ad esempio, il piÃ¹ basso) Ã¨ scelto come il miglior modello e viene ri-allenato stavolte sull'interezza dei dati (senza fare fold)

```python
# Function to perform the K-fold cross validation
def k_fold_cross_validation(x_train: np.ndarray, y_train: np.ndarray, k: int, max_degree: int) -> tuple[tuple, tuple]:
    # TODO: Perform K-fold cross-validation on the training set.
    # The two returned values are the best model and the list of results for all degrees up to max_degree.
    # The points should be selected randomly.
    # The inputs and labels are already in terms of distance and SNR

    # Divide training set in K folds:
    tot_points = len(x_train)
    fold_points = int( np.floor( tot_points/k ) )
    tot_idx = np.arange(tot_points)
    idx_mix = npr.permutation(tot_idx)
    x_folds = [ x_train[ idx_mix[ i*fold_points : (i+1)*fold_points ] ] for i in range(k) ]
    y_folds = [ y_train[ idx_mix[ i*fold_points : (i+1)*fold_points ] ] for i in range(k) ]


    # Train and validate:
    results=[]        # list of results for each degree
    best_mse=10**10   # mean mse used for searching best fit
    best_coeff=[]     # list of best results
    for i_deg in range(max_degree + 1):
        mse_tot = 0
        for i_fold in range(k):
            x_i_train = np.concatenate( [ x_folds[i] for i in range(k) if i_fold != i ] )
            y_i_train = np.concatenate( [ y_folds[i] for i in range(k) if i_fold != i ] )
            x_i_valid = x_folds[i_fold]
            y_i_valid = y_folds[i_fold]
            # training:
            coeff = fit(x_i_train, y_i_train, i_deg)
            # validation:
            mse_tot += evaluate(x_i_valid, y_i_valid, coeff)
        mean_mse = mse_tot/k
        results.append( [i_deg, mean_mse] )
        # if result is better than previous, we train it over all training set to obtain training error:
        if (mean_mse < best_mse):
            best_mse = mean_mse
            best_coeff = fit(x_train, y_train, i_deg)
    best_coeff_k_fold = best_coeff
            
    return best_coeff_k_fold, results
```

```python
# Run the training with K-fold cross-validation with 40 test points and k=4 folds 

# initialization as numpy array for errors:
x = np.array(df['x'])
y = np.array(df['y'])
SNR = np.array(df['SNR'])

# initial values:
test_points = 40
k = 4
max_degree = 6

# compute distance:
distance = compute_distance(x, y)
# separate data in training set adn test set:
x_train, y_train, x_test, y_test = separate_test(distance, SNR, test_points)
# run k-fold cross validation:
best_coeff_k_fold, results = k_fold_cross_validation(x_train, y_train, k, max_degree)
print(f"The best coefficient are for degree n={len(best_coeff_k_fold)-1} and are given by \n{best_coeff_k_fold}")


# Plot the validation score as a function of the degree:
x_deg = [ coppia[0] for coppia in results ]
y_valid_mse = [ coppia[1] for coppia in results ]
plt.figure(figsize=(7,4))
plt.plot(x_deg, y_valid_mse)
plt.xlabel('Degree of polynomium')
plt.ylabel('Mean Squared Error (MSE)')
```

## B. Tikhonov regularization

Change the loss function to include a Tikhonov regularization term, as an alternative to cross-validation (try $\lambda=0.01$)

```python
# this time without k-fold


def evaluate_tikhonov(x_train: np.ndarray, y_train: np.ndarray, lambda_par: float, max_degree: int) -> tuple[tuple, tuple]:
    # Apply Tikhonov regularization AFTER the fitting process:

    results = []
    best_coeff = []
    best_error=10**10
    for i_deg in range(max_degree+1):
        coeff = np.polyfit(x_train, y_train, i_deg)
        error = evaluate(x_train, y_train, coeff)
        tik_error = error + np.sum( coeff**2)*lambda_par
        results.append( [i_deg, tik_error] )
        if (tik_error < best_error):
            best_error = tik_error
            best_coeff = coeff
    best_coeff_tik = best_coeff

    return best_coeff_tik, results
```

```python
# Run the training with Tikhonov regularization and plot the loss as a function of the degree

lambda_par = 0.01
best_coeff_tik, results = evaluate_tikhonov(x_train, y_train, lambda_par, max_degree)

x_deg = [ coppia[0] for coppia in results ]
y_tik_err = [ coppia[1] for coppia in results ]
plt.figure(figsize=(7,4))
plt.plot(x_deg, y_tik_err)
plt.xlabel('Degree of polynomium')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('MSE with Tikhonov')
results
```

## C. Minimum description length regularization (MDL)

Change the loss function to include a representation length regularization term, as an alternative to cross-validation. The minimum description length of a polynomial of degree N is O(2^N) - try $\lambda=0.02$

```python
# this time also without k-fold

def evaluate_representation(x_train: np.ndarray, y_train: np.ndarray, lambda_par: float, max_degree: int) -> tuple[tuple, tuple]:
    # Apply MDL regularization AFTER the fitting process:
    
    results = []
    best_coeff = []
    best_error=10**10
    for i_deg in range(max_degree+1):
        coeff = np.polyfit(x_train, y_train, i_deg)
        error = evaluate(x_train, y_train, coeff)
        mdl_error = error + lambda_par*2**(i_deg+1)
        results.append( [i_deg, mdl_error] )
        if (mdl_error < best_error):
            best_error = mdl_error
            best_coeff = coeff
    best_coeff_mdl = best_coeff

    return best_coeff_mdl, results
```

```python
# Run the training with MDL regularization and plot the loss as a function of the degree:

lambda_par = 0.02
best_coeff_mdl, results = evaluate_representation(x_train, y_train, lambda_par, max_degree)

x_deg = [ coppia[0] for coppia in results ]
y_mdl_err = [ coppia[1] for coppia in results ]
plt.figure(figsize=(7,4))
plt.plot(x_deg, y_mdl_err)
plt.xlabel('Degree of polynomium')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('MSE with Minimum Description Length')
results
```
